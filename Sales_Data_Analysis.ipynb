### Data Cleaning and Wrangling

```python
## Merge with customers

df = df_sales.merge(
    df_customers,
    how='left',
    left_on='Customer Name Index',
    right_on='Customer Index'
    )

df.head(6)
```

```python

## Merge with products

df = df.merge(
    df_products,
    how='left',
    left_on='Product Description Index',
    right_on='Index'
    )

df.head(5)
```

```python

## Merge with regions
df = df.merge(
    df_region,
    how='left',
    left_on='Delivery Region Index',
    right_on='id'
    )

df.head(5)
```

```python

# Merge with State_regions

df = df.merge(
    df_state_region[['State Code', 'Region']],
    how='left',
    left_on='state_code',
    right_on='State Code'
    )

df.head(5)


# df_state_region.columns.values
```

```python

# Merge with budget

df = df.merge(
    df_budget,
    how='left',
    on='Product Name')

df.head(5)
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns.values
```

```python
# # Step 1: List of columns to keep (original names)
cols_to_keep = [
    'ordernumber', 'orderdate', 'customer names', 'channel',
    'product name', 'order quantity', 'unit price', 'line total',
    'total unit cost', 'county', 'state_code', 'state', 'region',
    'latitude', 'longitude', '2017 budgets'
]

# # Step 2: Filter the DataFrame to keep only relevant columns
df = df[cols_to_keep]

# Step 3: Rename the columns to clean snake_case names
df = df.rename(columns={
    'ordernumber': 'order_number',
    'orderdate': 'order_date',
    'customer names': 'customer_name',
    'channel': 'sales_channel',
    'product name': 'product_name',
    'order quantity': 'order_quantity',
    'unit price': 'unit_price',
    'line total': 'Revenue',
    'total unit cost': 'total_unit_cost',
    'county': 'county',
    'state_code': 'state_code',
    'state': 'state',
    'region': 'region',
    'latitude': 'latitude',
    'longitude': 'longitude',
    '2017 budgets': 'budget'
})
```

```python
df.columns.values
```

```python
df.loc[df['order_date'].dt.year != 2017, 'budget'] = pd.NA

df.head()
```

###Feature Engineering

```python
df['total_cost'] = df['order_quantity'] * df['total_unit_cost']
```

```python
df['profit'] = df['Revenue'] - df['total_cost']
```

```python
df['profit_margin_pct'] = df['profit'] / df['Revenue'] * 100
df.head()
```

```python
df['order_month'] = df['order_date'].dt.to_period('M')
monthly_sales = df.groupby('order_month')['Revenue'].sum()
plt.figure(figsize=(15,4))
monthly_sales.plot(marker='o',c='blue')

from matplotlib.ticker import FuncFormatter

formatter = FuncFormatter(lambda x, pos: f'{x/1e6:.1f}M')

plt.gca().yaxis.set_major_formatter(formatter)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue (Millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Ensure order_date is datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Create order_month column (if not already)
df['order_month'] = df['order_date'].dt.to_period('M')

# Group and aggregate revenue
monthly_sales = df.groupby('order_month')['Revenue'].sum().sort_index()

# Convert PeriodIndex to datetime for formatting
monthly_sales.index = monthly_sales.index.to_timestamp()

# Format x-axis as "Jan 2023", etc.
month_labels = monthly_sales.index.strftime('%b %Y')

# Plot
plt.figure(figsize=(15, 5))
plt.plot(month_labels, monthly_sales.values, marker='o', color='dodgerblue', linewidth=2)

# Format y-axis in millions
formatter = FuncFormatter(lambda x, pos: f'${x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)

# Add data labels
for x, y in zip(month_labels, monthly_sales.values):
    plt.text(x, y, f'{y/1e6:.1f}M', ha='center', va='bottom', fontsize=8, color='gray')

# Titles and styling
plt.title('Monthly Sales Trend by Revenue', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue (Millions)', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
monthly_margin = df.groupby('order_month')['profit_margin_pct'].mean().sort_index()
monthly_margin.index = monthly_margin.index.to_timestamp()

plt.figure(figsize=(15, 4))
plt.plot(monthly_margin.index.strftime('%b %Y'), monthly_margin.values, marker='s', color='green')
plt.title("Monthly Average Profit Margin (%)")
plt.ylabel("Profit Margin (%)")
plt.xlabel("Month")
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# top_products = df.groupby('product_name')['Revenue'].sum().nlargest(10)
# top_products.plot(kind='bar', color='royalblue')
# plt.title("Top 10 Products by Revenue")
# plt.ylabel("Total Revenue")
# plt.xlabel("Product")
# plt.tight_layout()
# plt.show()



# Step 1: Group by product and sum revenue
top_products = df.groupby('product_name')['Revenue'].sum().nlargest(10).reset_index()

# Step 2: Sort by revenue for better bar chart appearance
top_products = top_products.sort_values(by='Revenue', ascending=False)

# Step 3: Plot with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='Revenue', y='product_name', palette='viridis')

# Step 4: Add annotations
for i, value in enumerate(top_products['Revenue']):
    plt.text(value, i, f"${value/1e6:.2f}M", va='center', ha='left', fontsize=9)

# Styling
plt.title("Top 10 Products by Total Revenue", fontsize=16, fontweight='bold')
plt.xlabel("Total Revenue")
plt.ylabel("Product Name")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python


# Group revenue by sales channel
chan_sales = df.groupby('sales_channel')['Revenue'].sum().sort_values(ascending=False)

# Data
values = chan_sales.values
labels = chan_sales.index
colors = plt.cm.Paired(np.linspace(0, 1, len(labels)))  # modern color palette

# Optional: explode the largest slice
explode = [0.05 if i == 0 else 0 for i in range(len(labels))]

# Plot
fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts, autotexts = ax.pie(
    values,
    labels=None,  # we'll customize labels below
    autopct='%1.1f%%',
    startangle=120,
    pctdistance=0.85,
    explode=explode,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}
)

# Draw center circle for donut shape
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Custom legend (outside the plot)
ax.legend(wedges, labels, title="Sales Channels", loc="center left", bbox_to_anchor=(1, 0.5))

# Style adjustments
plt.setp(autotexts, size=10, weight='bold', color='black')
plt.title("Revenue Breakdown by Sales Channel", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#Average order value (AOV) Distribution

aov = df.groupby('order_number')['Revenue'].sum()
plt.figure(figsize=(12,4))
plt.hist(
    aov,
    bins =50,
    color='Skyblue'
)


plt.title('Average Order Value Distribution0')
plt.xlabel('Average order value in USD')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()

```

```python
#top 10 vs bottom 10 states by revenue
bottom_states = df.groupby('state')['Revenue'].sum().nsmallest(10)
top_states = df.groupby('state')['Revenue'].sum().nlargest(10)


# Convert to DataFrame for plotting
top_states_df = top_states.reset_index().sort_values(by='Revenue', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_states_df, x='Revenue', y='state', palette='Blues_d')

# Add value labels
for i, value in enumerate(top_states_df['Revenue']):
    plt.text(value, i, f"${value/1e6:.2f}M", ha='left', va='center', fontsize=9)

plt.title("Top 10 States by Total Revenue", fontsize=16, fontweight='bold')
plt.xlabel("Total Revenue")
plt.ylabel("State")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
bottom_states = df.groupby('state')['Revenue'].sum().nsmallest(10)
bottom_states_df = bottom_states.reset_index().sort_values(by='Revenue', ascending=True)


plt.figure(figsize=(12, 6))
sns.barplot(data=bottom_states_df, x='state', y='Revenue', palette='Reds_r')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Add value labels
for i, value in enumerate(bottom_states_df['Revenue']):
    plt.text(i, value + value * 0.01, f"${value/1e6:.2f}M", ha='center', fontsize=9)

# Styling
plt.title("Bottom 10 States by Total Revenue", fontsize=16, fontweight='bold')
plt.xlabel("State")
plt.ylabel("Total Revenue")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
